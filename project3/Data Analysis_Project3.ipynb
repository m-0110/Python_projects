```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

## Polynomial Regression

```python
def polynomial_regression_function(data,attr1,attr2,deg):
    x=data[attr1].values
    y=data[attr2].values
    poly=PolynomialFeatures(degree=deg)
    x_poly=poly.fit_transform(x.reshape(-1,1))
    poly_model=LinearRegression()
    poly_model.fit(x_poly,y.reshape(-1,1))
    y_pred=poly_model.predict(x_poly)
    error=np.sqrt(mean_squared_error(y,y_pred))
    plt.scatter(x,y)
    plt.plot(x,y_pred,c='red')
    plt.xlabel(attr1)
    plt.ylabel(attr2)
    plt.title('Regression Analysis')
    plt.xticks(rotation=90)
    plt.ticklabel_format(style='plain',axis='x')
    plt.ticklabel_format(style='plain',axis='y')
    #plt.savefig("F:\\Python_Track_Training\\Statistics\\project3\\" + attr1+"vs"+attr2+".png")
    print(poly_model.coef_)
    return (r2_score(y, y_pred),error)
    
```

## Multiple Regression

```python
def multiple_regression_function(data,attr1,attr2):
    x_data=data[attr1]
    y_data=data[attr2]
    linearmodel=LinearRegression()
    linearmodel.fit(x_data, y_data)
    
    y_predict = linearmodel.predict(x_data)

    print("Regression coefficient",linearmodel.coef_)
    print("Intercept",linearmodel.intercept_)
    print("r2_score",r2_score(y_data,y_predict))
```
